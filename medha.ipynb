```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path='datafile.csv'
df = pd.read_csv(file_path)
second_10_columns =df.iloc[:, 10:20]
new_data1=df['District Names']
second_10_columns['District Names'] = new_data1
column_name=['Deaths in the last 3 years registered with the civil authority (%)', 'Households with any usual member covered under a health insurance/financing scheme (%)','Children age 5 years who attended pre-primary school during the school year 2019-20 (%)']
df = second_10_columns.drop(column_name, axis=1)
grouped_data = df.groupby('District Names').mean()
grouped_data.reset_index(inplace=True, drop=False)
df_1=grouped_data.head(2)
df_1=df_1.T
df_1.reset_index(drop=False, inplace=True)
df_1= df_1.drop(0)
df_1['parameter'] = df_1['index']
df_1['Adilabad'] = df_1[0]
df_1['Agar Malwa'] = df_1[1]
column_name=['index', 0,1]
df_1 = df_1.drop(column_name, axis=1)
```

```python
df_1
```

```python
variables = df_1['parameter']
entity1_data = df_1['Adilabad'].astype(float)
entity2_data = df_1['Agar Malwa'].astype(float)
print(entity2_data[1])
```

```python
entity2_data
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for two entities and six variables
variables = df_1['parameter']
entity1_data = df_1['Adilabad']
entity2_data = df_1['Agar Malwa']

# Calculate angles for each variable (in radians) to evenly space the axes
angles = np.linspace(0, 2 * np.pi, len(variables), endpoint=False)

# Close the plot to form a closed polygon
entity1_data = np.concatenate((entity1_data, [entity1_data[1]]))
entity2_data = np.concatenate((entity2_data, [entity2_data[1]]))
angles = np.concatenate((angles, [angles[0]]))

# Plotting the spider chart
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

# Plot entity 1's data
ax.plot(angles, entity1_data, 'o-', label='Adilabad')
ax.fill(angles, entity1_data, alpha=0.65)

# Plot entity 2's data
ax.plot(angles, entity2_data, 'o-', label='Agar Malwa')
ax.fill(angles, entity2_data, alpha=0.65)

# Set the labels for each axis
ax.set_xticks(angles[:-1])
ax.set_xticklabels(variables)

# Add a legend
ax.legend(loc='upper right')

# Set the title
plt.title('Performance analysis.')

# Show the plot
plt.show()
print(entity1_data)
```

```python
# Replace 'data.xlsx' with the path to your Excel file
file_path = 'datafile.csv'
df = pd.read_csv(file_path)
data = df.to_numpy()
first_10_columns = df.iloc[:, :10]
column_name =['District Names','Sex ratio at birth for children born in the last five years (females per 1,000 males)']
df = first_10_columns.drop(column_name, axis=1)
grouped_data = df.groupby('State/UT').mean()
grouped_data.reset_index(inplace=True, drop=False)
```

```python
grouped_data
```

```python
df_grouped=grouped_data.head(2).T
df_grouped.reset_index(drop=False, inplace=True)
```

```python
df_grouped[0]
```

```python
# Drop the first row
df_grouped= df_grouped.drop(0)

# Reset the index after dropping the first row
df_grouped.reset_index(drop=True, inplace=True)
df_grouped['Andaman & Nicobar Islands'] = df_grouped[0]
df_grouped['Lakshadweep'] = df_grouped[1]
```

```python
df_grouped
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for two entities and six variables
variables = df_grouped['index']
entity1_data = df_grouped['Andaman & Nicobar Islands']
entity2_data = df_grouped['Lakshadweep']

# Calculate angles for each variable (in radians) to evenly space the axes
angles = np.linspace(0, 2 * np.pi, len(variables), endpoint=False)

# Close the plot to form a closed polygon
entity1_data = np.concatenate((entity1_data, [entity1_data[0]]))
entity2_data = np.concatenate((entity2_data, [entity2_data[0]]))
angles = np.concatenate((angles, [angles[0]]))

# Plotting the spider chart
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, polar=True)

# Plot entity 1's data
ax.plot(angles, entity1_data, 'o-', label='Andaman & Nicobar Islands')
ax.fill(angles, entity1_data, alpha=0.65)

# Plot entity 2's data
ax.plot(angles, entity2_data, 'o-', label='Lakshadweep')
ax.fill(angles, entity2_data, alpha=0.65)

# Set the labels for each axis
ax.set_xticks(angles[:-1])
ax.set_xticklabels(variables)

# Add a legend
ax.legend(loc='upper right')

# Set the title
plt.title('Performance analysis.')

# Show the plot
plt.show()
print(entity1_data)
```
